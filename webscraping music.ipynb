```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
```

```python
def music(url):
    
    lis=[]
    for i in range(0,len(url)):
        html= urlopen(url.iloc[i,0])
        soup = BeautifulSoup(html)
        song=soup.find_all("li",{"class":"s_title"})
        for x in song:
            lis.append(x.get_text())
            
    a=[f.split("  ") for f in lis]
    
    for i in a:   
        if i==['title']:
            a.remove(['title'])
            
    for i in a:
        while("" in i): 
            i.remove("")
    
    for i in range(len(a)):
        a[i]=[f for f in a[i] if a[i].index(f)<2]
        
    content=pd.DataFrame(a, columns =["song","artist_1"])
    
    det=[]
    for i in range(0,len(url)):
        html= urlopen(url.iloc[i,0])
        soup = BeautifulSoup(html)
        info=soup.find_all("div",{"class":"_t1"})
        for x in info:
            det.append(x.get_text())
            
    k=[t.split("|") for t in det]

    album=[]
    record=[]
    release=[]
    for i in range(len(k)):
        album_name=k[i][0].split('Released by')[0]
        record_name=k[i][0].split('Released by')[1]
        release_date=k[i][1]
        no_of_tracks=int(k[i][2][1:3])
        for j in range(no_of_tracks):
            album.append(album_name)
            record.append(record_name)
            release.append(release_date)
            
    content=pd.concat([content,pd.DataFrame(album,columns=["album"]),pd.DataFrame(record,columns=["record_name"]),pd.DataFrame(release,columns=["release_month"])],axis=1)

    for i in range(len(content)):
        content['language']='Kannada'
    
    for i in range(len(content)):
        content['languageID']='3'
    return(content)  
```

```python
def divide_chunks(l, n): 
      
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 
  
# How many elements each 
# list should have 
n = 1
  
```

```python
f18 = list(divide_chunks(klist18, n)) 
f18
```

```python
len(f18)
```

```python
f19 = list(divide_chunks(klist19, n)) 
f19
```

```python
pytrends = TrendReq(hl='IN', tz=330)
```

```python
t18=[]
for x in range(0,len(f18)):
    pytrends.build_payload(f18[x], cat=0, timeframe='2018-12-01 2018-12-31', geo='IN', gprop='youtube')
    t18.append(pytrends.interest_over_time())
```

```python
def prepro(t4):
    
    for i in range(len(t4)):
        if 'isPartial' in t4[i].columns:
            t4[i].drop(['isPartial'],axis=1,inplace=True)
        
    for i in range(0,len(t4)):
        if (t4[i].empty == True) in t4[i]:
            del(t4[i])
        
    for i in range(0,len(t4)):
        t4[i]=t4[i].transpose()
        
    f4=pd.DataFrame()
    for i in range(0,len(t4)):
        f4=pd.concat([f4,t4[i]],axis=0)
        
    f4['avg']=round(f4.mean(axis=1,skipna=True))
    return(f4)
```

```python
avg18=prepro(t18)
```
